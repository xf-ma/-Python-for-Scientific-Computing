```python
# Fixing random state for reproducibility
np.random.seed(19680801)

Nsteps, Nwalkers = 100, 250
t = np.arange(Nsteps)

# an (Nsteps x Nwalkers) array of random walk steps
S1 = 0.004 + 0.02*np.random.randn(Nsteps, Nwalkers)
S2 = 0.002 + 0.01*np.random.randn(Nsteps, Nwalkers)

# an (Nsteps x Nwalkers) array of random walker positions
X1 = S1.cumsum(axis=0)
X2 = S2.cumsum(axis=0)


# Nsteps length arrays empirical means and standard deviations of both
# populations over time
mu1 = X1.mean(axis=1)
sigma1 = X1.std(axis=1)
mu2 = X2.mean(axis=1)
sigma2 = X2.std(axis=1)

# plot it!
fig, ax = plt.subplots(1)
ax.plot(t, mu1, lw=2, label='mean population 1')
ax.plot(t, mu2, lw=2, label='mean population 2')
ax.fill_between(t, mu1+sigma1, mu1-sigma1, facecolor='C0', alpha=0.4)
ax.fill_between(t, mu2+sigma2, mu2-sigma2, facecolor='C1', alpha=0.4)
ax.set_title(r'random walkers empirical $\mu$ and $\pm \sigma$ interval')
ax.legend(loc='upper left')
ax.set_xlabel('num steps')
ax.set_ylabel('position')
ax.grid()
```

```python
t = np.arange(100)
xs = np.random.normal(1,0.5,size=(100,1000))
mu = xs.mean(axis=1)
sigma = xs.std(axis=1)

fig, ax = plt.subplots(2)
ax[0].plot(t, xs, lw=2)
ax[0].set_title(r'random samples')

ax[1].plot(t, mu, lw=2)
ax[1].fill_between(t, mu+sigma, mu-sigma, facecolor='C0', alpha=0.4)
ax[1].set_title(r'random samples $\mu$ and $\pm \sigma$ interval')

ax[1].set_xlabel('num steps')
ax[1].grid()

plt.show()
```

```python
# Example from http://www2.math.umd.edu/~jmr/241/surfaces.html

import k3d
from k3d.colormaps import matplotlib_color_maps

x = np.linspace(-5, 5, 100, dtype=np.float32)
y = np.linspace(-5, 5, 100, dtype=np.float32)

x, y = np.meshgrid(x, y)
f = np.sin(x)*np.cos(y)

plt_surface = k3d.surface(f,                                                
                          compression_level=9,
                          color_map=matplotlib_color_maps.Coolwarm_r,
                          attribute=f, color_range=[-1, 0.5])


plot = k3d.plot()
plot += plt_surface
plot.display()
```
